### Question 08

Create a Series of 50 people birthday year (1925 to 2025)

and set indexes with their gender (random Female/Male).

then, separate these 50 pople into 6 different series based on their generation:

* Alpha: Born 2013–2025.
* Z: Born 1997–2012 (Ages 14–29).
* Millennials: Born 1981–1996 (Ages 30–45).
* X: Born 1965–1980 (Ages 46–61).
* Baby Boomers: Born 1946–1964 (Ages 62–80).
* Silent: Born 1925–1945 (Ages 81–101). 

Finally, print number of Female and Male popele in each generation.

```python
birth_years = np.random.randint(1925, 2026, 50)
genders = np.random.choice(["Female", "Male"], 50)
people = pd.Series(birth_years, index=genders)
people
```

```python
def generation(year):
    if 2013 <= year <= 2025:
        return "Alpha"
    elif 1997 <= year <= 2012:
        return "Gen Z"
    elif 1981 <= year <= 1996:
        return "Millennials"
    elif 1965 <= year <= 1980:
        return "Gen X"
    elif 1946 <= year <= 1964:
        return "Baby Boomers"
    else:
        return "Silent"

generations = people.apply(generation)
result = pd.crosstab(generations, people.index)
result
```
